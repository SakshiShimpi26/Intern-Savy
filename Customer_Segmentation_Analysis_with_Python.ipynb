### **Plotting the distribution of gender to recognize different distributions**

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Gender', data=df);
plt.title('Distribution of Gender');
```

### **Creating a histogram of ages for given dataset.**

```python
plt.figure(figsize=(5,3))
df.hist('Age', bins=35);
plt.title('Distribution of Age');
plt.xlabel('Age');
```

### **From the above graph and the observation of describe() method we can notice that the average age of customer is 38.There are less older customers, so this distribution is right-skewed because of its longer right tail.**

```python
plt.figure(figsize=(5,3))
plt.hist('Age', data=df[df['Gender'] == 'Male'], alpha=0.5, label='Male');
plt.hist('Age', data=df[df['Gender'] == 'Female'], alpha=0.5, label='Female');
plt.title('Distribution of Age by Gender');
plt.xlabel('Age');
plt.legend();
```

### **From the above analysis we can say that the men in this data set tended to be younger than the women. You can see the spike around the age of 30â€“35 for the women is where the majority of them fall. There are also more middle-aged women in this data set than men. There is a significant amount of senior men in the 65â€“70 year old bucket.**

### **Plotting the graph for income.**

```python
plt.figure(figsize=(5,3))
df.hist('Annual Income (k$)');
plt.title('Annual Income Distribution in Thousands of Dollars');
plt.xlabel('Thousands of Dollars')
```

### **Here we can observe that much of the incomes lie between the 60 and 85,000 dollar buckets. So now we will check whether the gender affect the income?**

```python
# Histogram of income by gender
plt.hist('Annual Income (k$)', data=df[df['Gender'] == 'Male'], alpha=0.5, label='Male');
plt.hist('Annual Income (k$)', data=df[df['Gender'] == 'Female'], alpha=0.5, label='Female');
plt.title('Distribution of Income by Gender');
plt.xlabel('Income (Thousands of Dollars)');
plt.legend();
```

### **Creating data sets by gender to save time in the future since gender seems to significantly impact other variables.**

```python
male_customers = df[df['Gender'] == 'Male']
female_customers = df[df['Gender'] == 'Female']

# Print the average spending score for men and women
print("Male Customer Score:",male_customers['Spending Score (1-100)'].mean())
print("Female Customer Score:",female_customers['Spending Score (1-100)'].mean())
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(df.corr(), annot=True)
```

### **Hence we observed that there is no correlation between age and income**

### **Here we can see that the only variables that are even somewhat correlated is spending score and age. It is a negative correlation so the older a customer is in this data set, the lower their spending score.**

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age", y='Spending Score (1-100)',hue='Gender',data=df)
plt.title('Age to Spending Score, Colored by Gender');
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(female_customers.corr(), annot=True);
plt.title('Correlation Heatmap - Female');
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(male_customers.corr(), annot=True);
plt.title('Correlation Heatmap - Male');
```

### **From the above both graphs we can conclude that the age more strongly affects spending score for women in this case.**

### **Finding the womenâ€™s spending score to age relationship**

```python
plt.figure(figsize=(5,3))
sns.lmplot(data=female_customers,x='Age', y='Spending Score (1-100)');
plt.title('Age to Spending Score, Female Only');
```

### **Finding the income to spending score colored by gender**

```python
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', data=df);
plt.title('Annual Income to Spending Score, Colored by Gender');
```
